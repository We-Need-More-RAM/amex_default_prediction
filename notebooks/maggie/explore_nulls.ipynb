```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('numeric_vars_sample.csv')
```

```python
df = df.drop(columns=['S_2','Unnamed: 0', 'D_63', 'D_64', 'B_31', 'customer_ID'])
```

```python
for col in df.columns:
    df.loc[df[col] < -1, col] = -1
    
for col in df.columns:
    df.loc[df[col] > 10, col] = 10
    
df_stats = df.describe().T

fill_with_zero = df_stats[df_stats['min'] > 0].index
fill_with_neg2 = df_stats[df_stats['min'] <= 0].index

for col in fill_with_zero:
    df.loc[df[col].isnull(), col] = 0
    
for col in fill_with_neg2:
    df.loc[df[col].isnull(), col] = -2
```
