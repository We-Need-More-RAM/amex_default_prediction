## Catagorical Imputation Function

```python
def impute_categorical_columns(df):

    cat_columns = ['B_30', 'B_38', 'D_114', 'D_116', 'D_117', 'D_120', 'D_126', 'D_63', 'D_64', 'D_66', 'D_68']

    for column in cat_columns:

        if pd.api.types.is_numeric_dtype(df[column]):

            impute_value = df[column].value_counts().index.max() + 1

            df[column] = df[column].fillna(impute_value)

        else:

            impute_value = 'Unknown'

            df[column] = df[column].fillna(impute_value)
            
    return df
```

```python
impute_categorical_columns(X_df)
```
