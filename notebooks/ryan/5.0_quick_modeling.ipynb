```python
import ryan_src.eval as ev
import ryan_src.prep as prep

import pandas as pd

import warnings
warnings.filterwarnings("ignore")
```

```python
train, train_labels = prep.import_train()
val, val_labels = prep.import_validate()
```

```python
train = prep.prep_data(train)
train.to_csv('data/train_prep.csv', index=False)
```

```python
train
```

```python
from sklearn import tree
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(train, train_labels.target)
```

```python
test_preds = pd.DataFrame()
for i in range(0, 33):
    print(f'Starting file {i} process')
    df = pd.read_csv(f'data/chunked/test_data_chunk{i}.csv')
    df = prep.prep_data(df)
    
    # Fix missing columns
    
    for column in train.columns:
        if column not in df.columns:
            df[column] = 0
            
    # Removing columns that the model wasn't trained on
    
    for column in df.columns:
        if column not in train.columns:
            df = df.drop(columns=column)
    
    preds_from_df = clf.predict(df)
    preds_dict = {
        'customer_ID': df.index.tolist(),
        'preds': preds_from_df,
        }
    preds_df = pd.DataFrame.from_dict(preds_dict)
    test_preds = pd.concat([test_preds, preds_df], ignore_index=True)
```

```python
test_preds.to_csv('../../data/submission/first_test_submission.csv', index=False)
```
