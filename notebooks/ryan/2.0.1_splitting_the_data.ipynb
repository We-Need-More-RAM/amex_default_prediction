# Splitting the data

## Imports

```python
import pandas as pd

from sklearn.model_selection import train_test_split
```

## Bring in the data

```python
X_df = pd.read_csv('../../data/raw/train_data.csv')
X_df
```

```python
y_df = pd.read_csv('../../data/raw/train_labels.csv')
y_df
```

## Since this is longitudinal data we will use the train_labels for the split then get the correct data from train_data

```python
train_labels, val_test = train_test_split(y_df, train_size=0.085, random_state=13)
val_labels, test_labels = train_test_split(val_test, test_size=0.49, random_state=13)

print('Train: %d rows, %d cols' % train_labels.shape)
print('Validate: %d rows, %d cols' % val_labels.shape)
print('Test: %d rows, %d cols' % test_labels.shape)
```

```python
train_data = X_df[X_df.customer_ID.isin(train_labels.customer_ID.unique())]
val_data = X_df[X_df.customer_ID.isin(val_labels.customer_ID.unique())]
test_data = X_df[X_df.customer_ID.isin(test_labels.customer_ID.unique())]

print('Train: %d rows, %d cols' % train_data.shape)
print('Validate: %d rows, %d cols' % val_data.shape)
print('Test: %d rows, %d cols' % test_data.shape)
```
