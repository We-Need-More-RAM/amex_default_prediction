```python
#Import dependencies
import pandas as pd
import seaborn as sns
import datetime as dt
```

```python
#Declare the common path for the files, and read them in
path = '../../data/prepared/'

train = pd.read_csv(path + 'train_data.csv')
train_labels = pd.read_csv(path + 'train_labels.csv')

train.head()
```

```python
#Merge the labels to the features
train = train.merge(train_labels, how='left', on='customer_ID')

train.head()
```

```python
#Define a small function to return a datetime object from a string
def datetimer(string):
    return dt.datetime.strptime(string, '%Y-%m-%d')
```

```python
#Apply my function to the date column
train['S_2'] = train['S_2'].apply(datetimer)
```

```python
#Create a dataframe of the earliest date for each customer
min_dates = pd.DataFrame(train.groupby('customer_ID').S_2.min())

min_dates.head()
```

```python
#Rename the column to min_dates
min_dates.rename(columns={'S_2': 'min_dates'}, inplace=True)
```

```python
#Merge my min_dates column back to the main dataframe
train = train.merge(min_dates, how='left', on='customer_ID')
```

```python
#Calculate an age column, and return the number of days
train['age'] = (train['S_2'] - train['min_dates']).dt.days
```

```python
#Drop the columns that are no longer needed
train.drop(columns=['S_2', 'min_dates'], inplace=True)
```

```python
#Check my work was successful
train.head()
```

## Features to engineer for columns with clear patterns

P_4, D_42, D_49, D_53, D_55, D_61, D_75, D_134, D_137, B_16, B_20, R_2, R_4, R_5, R_6

P_4: many defaulting customers begin at 1 and end at 0, this feature is captured by the max of the feature

D_42: presence of values indicates default, encode binary column - if all values in column are null, return 0, else 1

D_49: same as above, encode binary column - if all values are null, return 0, else 1

D_53: higher values associated with default, positive change near end of account, take slope of recent 6-9 months

D_55: greater positive slopes for accounts that default, take most recent value minus initial value

D_61: higher values associated with default, nothing useful to encode

D_75: higher values associated with default, nothing useful to encode

D_134 through D_138: presence of values is indicative of default, binary column for all null = 0 else 1

B_16: greater positive slopes are associated with default, take slope of line and/or recent value minus initial value

B_19: large values associated with default, max() will capture this

B_20: greater positive slopes are associated with default, take slope of line and/or recent value minus initial value

R_2, R_4: having a value of 1 at any time is indicative of default, this will be captured by taking the max of the feature

R_5, R_6: having a value greater than 0 is indicative of default, this will be captured by the max of the feature

```python
#Create a dataframe with the sum of values in D_42
train_compressed = pd.DataFrame(train.groupby('customer_ID')['D_42'].sum())

train_compressed
```

```python
#Rewrite the D_42 to be a binary feature
train_compressed['D_42'].where(train_compressed['D_42'] == 0, other=1, inplace=True)
```

```python
#Check the feature was encoded successfully
train_compressed['D_42'].value_counts()
```

```python
#Rename the column so I don't run into issues later
train_compressed.rename(columns = {'D_42': 'D_42_has_value'}, inplace=True)
```

```python
#Encode another binary feature if D_49 has all nulls or not
train_compressed['D_49_has_value'] = train.groupby('customer_ID')['D_49'].sum()
train_compressed['D_49_has_value'].where(train_compressed['D_49_has_value'] == 0, other=1, inplace=True)
```

## Has value features

These features are engineered based on visuals generated in the explore notebook. Accounts that defaulted were more likely to contain values in these columns compared to accounts that did not. I engineered a binary feature to capture this difference: accounts with any values in these columns receive a 1 for the new feature, and accounts with all null values receive a 0.

These features were also engineered for a number of columns below: D_134 through D_138.

```python
#Check my work
train_compressed.value_counts()
```

```python
#Use boolean masking to extract values from the initiation of the accounts
train[train['age'] == 0]
```

```python
#Capture the maximum age for each customer
train.groupby('customer_ID')['age'].max()
```

```python
#Generate a boolean mask for maximum age
train.groupby('customer_ID')['age'].transform(max) == train['age']
```

```python
#Use my boolean mask on a column of interest
train[train.groupby('customer_ID')['age'].transform(max) == train['age']]['D_55']
```

```python
#Use the same boolean masking technique for values at the beginning of an account
train[train.groupby('customer_ID')['age'].transform(min) == train['age']]['D_55']
```

```python
#Create a new column of the difference between D_55 at the end and beginning of an account
train_compressed['D_55_change'] = train[train.groupby('customer_ID')['age'].transform(max) == train['age']]['D_55'].to_numpy() - train[train.groupby('customer_ID')['age'].transform(min) == train['age']]['D_55'].to_numpy()
```

## Change features

The change features capture the change in a variable over the lifespan of the account. It's calculated by subtracting the initial value for a feature from the final value. Accounts that defaulted showed a clear positive trend in these features over time. Accounts that did not default had weak positive or neutral trends. By subtracting the initial value from the final one, I am able to capture this change in value over time.

Applied this analysis to B_16 and B_20 below.

```python
#Create a list of columns I wish to encode in a similar manner
del_binary_encode = ['D_134', 'D_135', 'D_136', 'D_137', 'D_138']
```

```python
#Use a basic for loop to encode the "has_value" feature for my columns of interest
for col in del_binary_encode:
    train_compressed[col + '_has_value'] = train.groupby('customer_ID')[col].sum()
    train_compressed[col + '_has_value'].where(train_compressed[col + '_has_value'] == 0, other=1, inplace=True)
```

```python
#Check my work
train_compressed.head()
```

```python
#Create a looping function to calculate the change in values over the lifespan of an account
b_slope_encode = ['B_16', 'B_20']

for col in b_slope_encode:
    train_compressed[col + '_change'] = train[train.groupby('customer_ID')['age'].transform(max) == train['age']][col].to_numpy() - train[train.groupby('customer_ID')['age'].transform(min) == train['age']][col].to_numpy()
```
