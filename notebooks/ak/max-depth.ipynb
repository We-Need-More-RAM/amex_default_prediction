```python
import pandas as pd
import numpy as np
import xgboost as xgb
import myfunctions as mf
from sklearn.model_selection import train_test_split
```

```python
seed = 42
```

```python
df = pd.read_csv('prepped_data.csv')

train_labels = pd.read_csv('../../data/prepared/train_labels.csv')
val_labels = pd.read_csv('../../data/prepared/val_labels.csv')
```

```python
all_labels = pd.concat([train_labels, val_labels], axis=0)
```

```python
all_labels
```

```python
df = df.merge(all_labels, how='left', on='customer_ID')
```

```python
df
```

```python
train, val = train_test_split(df, train_size=0.9, random_state=seed)
```

```python
val
```

```python
train_matrix = xgb.DMatrix(train.drop(columns=['customer_ID', 'target']), label=train['target'])
val_matrix = xgb.DMatrix(val.drop(columns=['customer_ID', 'target']), label=val['target'])
```

```python
params = {
    'max_depth': 5,
    'learning_rate': 0.075,
    'subsample': 0.8,
    'colsample_bytree': 0.8,
    'eval_metric': 'logloss',
    'objective': 'binary:logistic',
    'random_state': seed
}
```

```python
model = xgb.train(params, dtrain=train_matrix,
                  evals=[(train_matrix, 'train'), (val_matrix, 'val')],
                  num_boost_round=9999,
                  early_stopping_rounds=5,
                  verbose_eval=25)
```

```python
score = mf.model_evaluator(model, val_matrix, val['target'])
```

```python
score
```
