```python
#Import dependencies
import pandas as pd
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Define the path and declare the complete paths for data/labels
train_path = '../../data/prepared/'

train_data_path = train_path + 'train_data.csv'
train_labels_path = train_path + 'train_labels.csv'
```

```python
#Read in the data
df = pd.read_csv(train_data_path)

df.head()
```

```python
#Create dataframe of the categorical variables
df_cat = df[['customer_ID', 'B_30', 'B_38', 'D_114', 'D_116', 'D_117', 'D_120', 'D_126', 'D_63', 'D_64', 'D_66', 'D_68']]

df_cat.head()
```

```python
#Read in the labels
df_labels = pd.read_csv(train_labels_path)

df_labels.head()
```

```python
#Merge the dataframes on customer ID
df_train = pd.merge(df_cat, df_labels, on='customer_ID')

df_train
```

```python
df_train.info()
```

```python
#Create a list of the columns
column_list = list(df_train.columns)

column_list
```

```python
#Remove the column names not associated with categorical variables
column_list.remove('customer_ID')
column_list.remove('target')

column_list
```

```python
#Print the value counts for all categorical values
for col in column_list:
    print('Column name: {}.\n'.format(col))
    print(df_train[col].value_counts())
    print()
```

```python
for value in df_train['B_30'].value_counts().index:
    print(value)
    print(df_train[df_train['B_30'] == value].target.value_counts())
```

```python
pd.crosstab(df_train.B_30, df_train.target)
```

```python
corr = {}

for var in column_list:
    observed = pd.crosstab(df_train[var], df_train['target'])
    print(observed)
    chi2, p, degf, expected = stats.chi2_contingency(observed)
    corr[var] = chi2, p
```
